```python
#if

salary = 10000
if salary > 8000:
  print('I will move to usa')
```

```python
#if else

salary = 10000
if salary < 8000:
  print('I will move to usa')
else:
  print('I will stay in india')
```

```python
# elif - meanse used to write multiple if condition

salary = 10000
if salary > 8000:
  print('I will move to usa')
else:
  print('I will stay in india')
```

```python
salary = 10000
if salary > 8000:
  print('I will move to usa')
elif salary > 900:
  print('I am ready')
else:
  print('I will stay in india')
```

```python
# even/odd
for i in range(1,11):
  if i %2==0:
    print('even- ',i)
  else:
    print('odd-  ',i)
```

```python
# even/odd
for i in range(1,11):
  if i %2==0:
    print('even- ',i)
```

```python
# even/odd
for i in range(1,11):
  if i %2 !=0:
    print('odd- ',i)
```

```python
# break - trying to stop something

for i in range (41,51):
  print(i)
  if i == 45:
    break
```

```python
# continue - meanse continue the loop
for i in range(10):
  if i > 4:
    print('hi all')
```

```python
# continue - meanse continue the loop
for i in range(10):
  if i > 4:
    print('hi all')
    continue
  elif i < 7:
    print(i)
```

```python
for i in range(50,60):
  if i == 56:
    continue
  print(i)
```

```python
for i in range(50,60):
  if i == 56:
    continue
  print(i,end = " ")
```
